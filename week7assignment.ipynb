```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Task 1: Load and Explore the Dataset
try:
    # Load the dataset
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = iris.target
    df['species'] = df['species'].map(dict(zip(range(3), iris.target_names)))

    print("âœ… Dataset loaded successfully.\n")
    print("ðŸ“Œ First five rows of the dataset:")
    print(df.head())

except FileNotFoundError:
    print("âŒ File not found.")
except Exception as e:
    print(f"âŒ An error occurred: {e}")

# Explore the structure of the dataset
print("\nðŸ“Œ Data types:")
print(df.dtypes)

print("\nðŸ“Œ Missing values:")
print(df.isnull().sum())

# Since there are no missing values, we don't need to clean the dataset

# Task 2: Basic Data Analysis
print("\nðŸ“Š Basic Statistics:")
print(df.describe())

# Grouping by species and getting mean of each feature
print("\nðŸ“Š Average features per species:")
grouped = df.groupby('species').mean()
print(grouped)

# Task 3: Data Visualization
sns.set(style="whitegrid")

# 1. Line Chart - Mean feature per species
grouped.T.plot(kind='line', marker='o')
plt.title("Line Chart: Average Feature per Species")
plt.xlabel("Feature")
plt.ylabel("Mean Value")
plt.legend(title="Species")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Bar Chart - Average petal length per species
sns.barplot(x='species', y='petal length (cm)', data=df)
plt.title("Bar Chart: Avg Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.tight_layout()
plt.show()

# 3. Histogram - Sepal Length Distribution
plt.hist(df['sepal length (cm)'], bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram: Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# 4. Scatter Plot - Sepal Length vs Petal Length
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df)
plt.title("Scatter Plot: Sepal vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.tight_layout()
plt.show()
```
