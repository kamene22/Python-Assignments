```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

```python
# 1.Creating a NumPy array from 1 to 10 and calculating the mean
array = np.arange(1, 11)
mean_value = np.mean(array)
print("Mean of array 1 to 10:", mean_value)
```

```python
#2.Loading a small dataset into a pandas DataFrame and displaying summary statistics

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 35, 40],
    'Score': [88, 92, 85, 95]
}
df = pd.DataFrame(data)
print("\nDataFrame Summary Statistics:")
print(df.describe())
```

```python
import requests

response = requests.get("https://api.exchangerate-api.com/v4/latest/USD")  
if response.status_code == 200:
    data = response.json()
    kes_rate = data["rates"].get("KES")
    if kes_rate:
        print("\nCurrent USD to KES Exchange Rate:", kes_rate)
    else:
        print("\nKES rate not found in the response.")
else:
    print("\nFailed to fetch exchange rate data.")
```

```python
# 4 Ploting a simple line graph using matplotlib
numbers = [1, 2, 3, 4, 5]
squares = [x**2 for x in numbers]
plt.plot(numbers, squares, marker='o', color='blue')
plt.title("Square Numbers")
plt.xlabel("Number")
plt.ylabel("Square")
plt.grid(True)
plt.show()
```
